# IPL EDA Python

### Importing Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np
```

```python
match_data=pd.read_csv("M:\Data SCience\IPL Matches.csv")
```

```python
match_data.head()
```

```python
match_data.info()
```

```python
match_data.drop(['method'],axis=1,inplace=True)
```

```python
# droping "method" feature cause of high nulls
plt.xkcd()
temp=pd.DataFrame({"Winner":match_data['winner']})
total_wins=temp.value_counts()
labels=[X[0] for X in total_wins.keys()]
bar,ax =plt.subplots(figsize=(20,20))
ax=plt.pie(x=total_wins,autopct='%.f%%',labels=labels)

plt.title("Most Wins In IPL History",fontsize=25)
plt.show()
```

```python
total_wins
```

# Most Wins in Eliminator

```python
sns.countplot(match_data['winner'][match_data['eliminator']=='Y'],data=match_data)
plt.xticks(rotation=90)
plt.xlabel("Teams in IPL")
plt.ylabel("No of Wins")
plt.title("Most Wins in Eliminator",fontsize=12)
plt.show()
```

```python
sns.countplot(match_data['player_of_match'][match_data['winner']=='Mumbai Indians'],data=match_data)
plt.xlabel=("Venues")
plt.ylabel("Total Wins")
plt.title("Total Wins of MUMBAI INDIANS in different venues")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Total Wins of top 4 IPL teams in different Venues

```python
sns.countplot(match_data['venue'][match_data['winner']=='Chennai Super Kings'],data=match_data)
plt.xlabel=("Venues")
plt.ylabel("Total Wins")
plt.title("Total Wins of Chennai Super Kings in different venues")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(match_data['venue'][match_data['winner']=='Kolkata Knight Riders'],data=match_data)
plt.xlabel=("Venues")
plt.ylabel("Total Wins")
plt.title("Total Wins of Kolkata Knight Riders in different venues")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(match_data['venue'][match_data['winner']=='Royal Challengers Bangalore'],data=match_data)
plt.ylabel("Total Wins")
plt.title("Total Wins of Royal Challengers Bangalore in different venues")
plt.xticks(rotation=90)
plt.show()
```

# Total wins of each team after wining toss

```python
sns.countplot(match_data['winner'][match_data['toss_winner']==match_data['winner'] ],data=match_data)
plt.xticks(rotation=90)
plt.ylabel("No of Wins")
plt.title("Total Wins after Winning TOSS",fontsize=12)
plt.show()
```

# TOP 5 UMPIRES 1

```python
sns.barplot(x=match_data['umpire1'].value_counts().head().index,
            y=match_data['umpire1'].value_counts().head().values,
            data=match_data
)
plt.xticks(rotation=90)
plt.show()
```
